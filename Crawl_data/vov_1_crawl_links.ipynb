```python
import requests
from bs4 import BeautifulSoup

BASE_URL = "https://vov.vn"

def get_article_links(page):

    url = f"{BASE_URL}/thi-truong/gia-ca-phe?page={page}"

    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, "html.parser")

    links = []

    for a in soup.find_all("a", href=True):

        href = a["href"]

        if "gia-ca-phe" in href and "post" in href:
            full_link = BASE_URL + href
            links.append(full_link)

    return list(set(links))
```

```python
all_links = []

for page in range(1,40):

    links = get_article_links(page)

    print(f"Page {page} - {len(links)} bài")

    all_links.extend(links)

all_links = list(set(all_links))

print("Tổng số bài:", len(all_links))
```

```python
import pandas as pd

df = pd.DataFrame({"url": all_links})

df.to_csv("vov_coffee_links.csv", index=False)
```
